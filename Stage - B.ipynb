1. From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
x = np.array((data['T2'].values)).reshape(-1,1)
y= np.array((data['T6'].values)).reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
model = reg.fit(x,y)
from sklearn.metrics import r2_score
r2 = model.score(x,y)
print(f'R2 score of T2 and T6 is {(round(r2,2))}')
```

2.  Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?

```python
df = data.drop(['date','lights'], axis =1)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
X = normalised_data.drop('Appliances', axis = 1)
y = normalised_data['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model
linear_model = linear_model.fit(x_train,y_train)
```

```python
#obtain predictions
predictions = linear_model.predict(x_test)
#MAE
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, predictions)
print(f'The Mean Absolute Error is {(round(MAE,2))}')
```

3. What is the Residual Sum of Squares (in two decimal places)?

```python
RSS = np.sum((predictions - y_test)**2)
print(f'The Residual SUm of Squares is {(round(RSS,2))}')
```

4. What is the Root Mean Squared Error (in three decimal places)?

```python
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y_test, predictions))
print(f'The Root Mean Squared Error is {(round(RMSE,2))}')
```

5. What is the Coefficient of Determination (in two decimal places)?

```python
r2 = r2_score(y_test, predictions)
print(f'The Coefficient of determination (r2_score) is {(round(r2,2))}')
```

6. Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_weights_data(model, feat, col_name):
    #this function returns the weight of every features
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_data = pd.DataFrame(weights).reset_index()
    weights_data.columns =['X', col_name]
    weights_data[col_name].round(3)
    return weights_data
linear_model_weights = get_weights_data(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights
```

```python
print(f'The feature with the lowest coefficient is RH_2: {round(linear_model_weights.Linear_Model_Weight.min(),2)}')
```

```python
print(f'The feature with the Highest coefficient is RH_1: {round(linear_model_weights.Linear_Model_Weight.max(),2)}')
```

7. Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
from sklearn.linear_model import Ridge
Ridge_Reg = Ridge(alpha = 0.5)
Ridge_Reg.fit(x_train, y_train)
y_pred = Ridge_Reg.predict(x_test)
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'The Root Mean Squared Error of Ridge Regreaaion is {(round(RMSE,2))}')
```

There is no change in the Root Mean Squared Error as both cases RMSE = 0.09

8. Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
from sklearn.linear_model import Lasso
Lasso_Reg = Lasso(alpha = 0.001)
Lasso_Reg.fit(x_train, y_train)
lasso_reg_weights = get_weights_data(Lasso_Reg, x_train, 'Lasso_Weights')
weights = lasso_reg_weights['Lasso_Weights']
for weight in weights:
    if weight != 0:
        print(weight)
```

Only 4 of the features have non zero feature weights
